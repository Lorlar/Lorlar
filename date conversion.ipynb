```python
#to creat a function 
def formatTime(estLectureTime, timeDiff=0):
    timeApart= estLectureTime.split(':')
    hr=int(timeApart[0])
    min=int(timeApart[1])
    if (hr-timeDiff < 12):
        return (str(hr-int(timeDiff)) + ":" + str(min).rjust(2, 'O') + 'AM')
    elif(hr - timeDiff == 12):
        return (str(hr-int(timeDiff)) + ":" + str(min).rjust(2, 'O') + 'PM')
    else:
        return(str(hr-int(timeDiff)-12) + ":" + str(min).rjust(2, 'O') + 'PM')
```
